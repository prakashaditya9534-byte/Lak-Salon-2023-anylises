```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Lakme 2023 data.csv")
```

```python
df = df.drop(columns=["Unnamed: 25","Unnamed: 26","Unnamed: 27"])
```

```python
df=df.fillna("Na")
```

```python
df.dtypes
```

```python
df["Booked Date"] = pd.to_datetime(df["Booked Date"], errors="coerce")
df["Sale Date"] = pd.to_datetime(df["Sale Date"], errors="coerce")
```

```python
df.dtypes
```

```python
num_cols = [
    "Price", "Discount", "LoyaltyPoint Redemption", "Net Price",
    "Tax", "Price Paid", "Sale Value", "PackageDiscount"]
```

```python
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
df.dtypes
```

```python
bool_map = {"Yes": True, "No": False, 1: True, 0: False}
df["FirstVisit"] = df["FirstVisit"].map(bool_map)
df["Member"] = df["Member"].map(bool_map)
```

```python
df.dtypes
```

Performance Analysis

```python
#Total Revenue
total_revenue = df['Sale Value'].sum()
print("Total Revenue:", total_revenue)
```

```python
#Monthly Revenue Trend
df['Month'] = df['Sale Date'].dt.to_period('M')

monthly_revenue = df.groupby('Month')['Sale Value'].sum()

print(monthly_revenue)
```

```python
#Plot Revenue Trend
monthly_revenue.plot(kind='line')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
#Revenue by Experts
revenue_by_Experts = df.groupby('Serviced By')['Sale Value'].sum().sort_values(ascending=False)

print(revenue_by_Experts)
```

Customer Analysis

```python
#Total Customers
total_customers = df['GuestCode'].nunique()

print("Total Customers:", total_customers)
```

```python
#Customer Visits
customer_visits = df.groupby('GuestCode')['Ticket No'].count()

print(customer_visits.head())
```

```python
#Define churn = No visit in last 90 days
last_visit = df.groupby('GuestCode')['Sale Date'].max().reset_index()

today = df['Sale Date'].max()

last_visit['days_since_last_visit'] = (today - last_visit['Sale Date']).dt.days
```

```python
#Identify churn customers
churn_customers = last_visit[last_visit['days_since_last_visit'] > 90]

print("Churn Customers:", churn_customers.shape[0])
```

```python
#Churn Rate
churn_rate = (len(churn_customers) / total_customers) * 100

print("Churn Rate:", churn_rate)
```

Top Services Analysis

```python
#Top Services by Revenue
top_services_revenue = df.groupby('ServiceName')['Sale Value'].sum().sort_values(ascending=False)

print(top_services_revenue.head(10))
```

```python
#Top Services by Demand
top_services_count = df.groupby('ServiceName')['ServiceName'].count().sort_values(ascending=False)

print(top_services_count.head(10))
```

```python
#Average Service Price
avg_service_price = df.groupby('ServiceName')['Sale Value'].mean().sort_values(ascending=False)

print(avg_service_price.head(10))
```

```python
#Employee Performance
employee_performance = df.groupby('Serviced By')['Sale Value'].sum().sort_values(ascending=False)

print(employee_performance.head(10))
```

```python
#Member vs Non-Member Revenue
member_analysis = df.groupby('Member')['Sale Value'].agg(['sum','count'])

print(member_analysis)
```

```python
#First Visit Customers
first_visit = df[df['FirstVisit'] == 'Yes']

new_customers = first_visit['GuestCode'].nunique()

print("New Customers:", new_customers)
```

```python
#Promotion Analysis
promotion_revenue = df.groupby('Promotion')['Sale Value'].sum().sort_values(ascending=False)

print(promotion_revenue)
```
